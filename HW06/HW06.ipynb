## Q2

```python
def data2(date,stockNo):
    info = {}
    info['date'] = date
    info['stockNo'] = stockNo
    url = "http://www.twse.com.tw/exchangeReport/STOCK_DAY?response=html&date={date}&stockNo={stockNo}".format(**info)
    r = requests.get(url)
    soup = BeautifulSoup(r.text ,"html.parser")
    soup_data = pd.read_html(str(soup.select('table')[0]),header=1)[0]
    return soup_data
```

```python
# 輸入日期及股票代碼，即可爬下當月的歷史資料
data2('20171130','2330')
```

## Q3

```python
def getdata(stock):
    info={}
    info['time'] = int(time.time() * 1000 + 1000000)
    info['stock'] = stock
    URL = 'http://mis.twse.com.tw/stock/api/getStockInfo.jsp?_={time}&ex_ch=tse_{stock}.tw'.format(**info)
    
    try:
    # Get original page to get session
        r = requests.session()
        r.get('http://mis.twse.com.tw/stock/index.jsp', headers={'Accept-Language': 'zh-TW'})

        response = r.get(URL)
        content = json.loads(response.text)
        
    except Exception as err:
        print(err)
        
        
    else:
        s = pd.DataFrame(content['msgArray'])
        data3 = pd.concat([s['c'], s['t'], s['z'], s['tv'], s['v']], axis=1)
        data3.columns =['股票編號', '資料時間', '最近價格', '成交量', '累計成交量']
        data3
        

    return data3
```

```python
getdata('2330')
```
