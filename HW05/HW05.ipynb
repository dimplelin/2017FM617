# Q1. 寫一個函式，抓取 Google Finance 的股價歷史資料

```python
import requests
import pandas as pd
```

```python
def gucha(stock,startdate):
    tuansu = {}
    tuansu['q'] = stock
    tuansu['startdate'] = startdate
    url = "https://finance.google.com/finance/historical?q={q}&startdate={startdate}&output=csv".format(**tuansu)
    stockprice=pd.read_csv(url)
    print(stockprice)

gucha('AAPL',"2016-01-01")
```
